```python
import pandas as pd
import numpy as np

pd.set_option('display.max.columns', 60)
```

```python
seq2 = ['A','C','C','G','A','C','T','A','G','A','C','A','G','G','T']
seq1 =  ['T','T','A','C','C','G','G','A','C','T','A','G','A','C','C','A','G','C','G','T','A']
```

```python
indel = ['_']

idt = 1
sub = -1
gap = -2
```

## programmation dynamique globale:


```python
def alignement_global(seqA,seqB):
    
    mat =  A = np.random.randint(1, size=(len(seqB)+1,len(seqA)+1)) #initialiser une matrice de zeros
    df = pd.DataFrame(data=mat,index=indel+seqB,columns=indel+seqA) #créer un tableau d'affichage
 
    for i in range(1,len(mat)) : 
        mat[i,0] = int(mat[i-1,0] + gap) #initialiser première ligne
    
    for j in range(1,len(mat[0])) :
        mat[0,j] = int(mat[0,j-1] + gap) #initialiser première colonne
    
    for i in range(1,len(mat)) :
        for j in range(1,len(mat[0])) :
        
            if seqB[i-1] == seqA[j-1] :
                #opération correspondante à deux bases identiques 
                mat[i,j] = max(mat[i,j-1] + gap, mat[i-1,j] + gap, mat[i-1,j-1] + idt)
           
            else :
                #opération correspondante à deux bases différentes 
                mat[i,j] = max(mat[i,j-1] + gap, mat[i-1,j] + gap, mat[i-1,j-1] + sub)
                
    score = mat[len(mat)-1,len(mat[0])-1] 
    print('le score =',score) #calaculer et afficher le score 
    
    
    #retour arrière
    x = len(mat)-1
    y = len(mat[0])-1

    horizontale = [seqA[y-1]]
    verticale = [seqB[x-1]]
    
    while(x > 1 and y > 1) :
        m = -999
        #détermination de la valeur maxx entre la case de gauche, dessus et diagonale

        if mat[x-1,y-1] > m : #diagonale
            x_max,y_max = x-1,y-1
            m = mat[x-1,y-1]

        if mat[x-1,y] > m : #gauche
            x_max,y_max = x-1,y
            m = mat[x-1,y]

        if mat[x,y-1] > m : #dessus
            x_max,y_max = x,y-1

        #insertion en tete des listes selon la position de la valeure max 
        if x_max == x : #case de gauche maximale (horizental)
    
            horizontale.insert(0,seqA[y_max-1])
            verticale.insert(0,indel[0]) # '_'

    

        if y_max == y : #case de dessus maximale (vertical)
            horizontale.insert(0,indel[0]) # '_'
            verticale.insert(0,seqB[x_max-1])
        
        if x_max == x-1 and y_max == y-1 : #case de diagonale maximale
        
            horizontale.insert(0,seqA[y_max-1])
            verticale.insert(0,seqB[x_max-1])


        x = x_max 
        y = y_max 

        ali = pd.DataFrame(data=[horizontale,verticale],index=['seq horizontale :', 'seq verticale :'])
    
    print('score daligenemnt = ', x+y)
    
    return df,ali
```

```python
df,ali = alignement_global(seq1,seq2)
```

```python
def local_align(x, y):
   
    A = []
    for i in range(len(y) + 1):
        A.append([0] * (len(x) +1))
    best = 0
    optloc = (0,0)

    for i in range(1, len(y) + 1):
        for j in range(1, len(x) + 1):
           
            A[i][j] = max(
            A[i][j-1] + gap,
            A[i-1][j] + gap,
            A[i-1][j-1] + idt if y[i-1] == x[j-1] else sub)
            
           
            if A[i][j] >= best:
                best = A[i][j]
                optloc = (i,j)
    
    align_X = " "
    align_Y =  " "
    j, i = optloc
    while (i > 0 or j > 0) and A[j][i] > 0:
         
        current_score = A[j][i]

        if i > 0 and j > 0 and x[i - 1] == y[j - 1]:
            align_X = x[i - 1] + align_X
            align_Y = y[j - 1] + align_Y
            i = i - 1
            j = j - 1
         
        elif i > 0 and (current_score == A[j][i - 1] + sub or current_score == A[j][i - 1] + gap) and A[j][i - 1] > 0:
            align_X = x[i - 1] + align_X
            align_Y = "-" + align_Y
            i = i - 1
             
        else:
            align_X = "_" + align_X
            align_Y = y[j - 1] + align_Y
            j = j - 1
            
        ali = pd.DataFrame(data=[align_X, align_Y],index=['seq horizontale :', 'seq verticale :'])

     
    return ali
```
